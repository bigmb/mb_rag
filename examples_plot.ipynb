```python
from mb_rag.utils.bounding_box import google_model,generate_bounding_box
from mb_rag.chatbot.basic import load_env
```

```python
import os
load_env('./env')
```

```python
model = google_model(model="gemini-1.5-pro-latest",api_key=os.environ.get("GOOGLE_API_KEY"))
```

```python
res =generate_bounding_box(model,image_path="/home/malav/Downloads/boilled_eggs.jpeg",prompt='Return bounding boxes of container, for each only one return [ xmin,ymin, xmax, ymax]')
```

```python
import json 
res = json.loads(res.text)    
res['container']
```

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread("/home/malav/Downloads/boilled_eggs.jpeg")
img = cv2.resize(img,(1000,1000))
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
cv2.rectangle(img, (res['container'][1],res['container'][0]), ( res['container'][3],res['container'][2]), (0, 0, 255), 4)
plt.imshow(img)
plt.show()


```
